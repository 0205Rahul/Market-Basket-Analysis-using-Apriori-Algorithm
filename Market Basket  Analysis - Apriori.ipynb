### Arranging the rules in descending order of lift values

```python
rules = apriori(df_trans, min_support = 0.003, min_confidence = 0.35, min_lift = 3, min_length = 2)
association_results = list(rules)
print(type(rules))
print(type(association_results[0]))
print(type(association_results))
print(association_results[0])
```

```python
len(association_results)
```

```python
store_data = pd.DataFrame(columns=('items','item1','item2','support','confidence','lift'))

support =[]
confidence = []
lift = []
items = []
item1 = []
item2=[]

for x in association_results:
    for stat in x.ordered_statistics:
        support.append(x.support)
        items.append(x.items)
        item1.append(stat.items_base)
        item2.append(stat.items_add)
        confidence.append(stat.confidence)
        lift.append(stat.lift)

store_data['items'] = list(map(set, items))                                   
store_data['item1'] = list(map(set, item1))
store_data['item2'] = list(map(set, item2))
store_data['support'] = support
store_data['confidence'] = confidence
store_data['lift']= lift
sortd=store_data.sort_values(by='lift',ascending=False)
```
